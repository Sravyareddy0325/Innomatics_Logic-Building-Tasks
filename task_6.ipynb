```python
def post_engagement(likes):
    total_likes = sum(likes)
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

post_engagement([200, 300, 250, 300])
```

```python
def medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    print(f"Medicine Stock: {stock}")
    print(f"Status: {status}")

medicine_stock(6)
```

```python
def rainfall_checker(rainfall, required=70):
    avg = sum(rainfall) / len(rainfall)
    status = "Adequate Rainfall" if avg >= required else "Inadequate Rainfall"
    print(f"Average Rainfall: {int(avg)}")
    print(f"Rainfall Status: {status}")

rainfall_checker([70, 75, 80, 65, 70])
```

```python
def duplicate_accounts(usernames):
    has_duplicates = len(usernames) != len(set(usernames))
    print("Duplicate Accounts Found:", "Yes" if has_duplicates else "No")

duplicate_accounts(["user1", "user2", "user1"])
```

```python
def appointment_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")

appointment_eligibility(21)
```

```python
def premium_crops(prices):
    premium = [p for p in prices if p > 2000]
    print(f"Premium Crops: {premium}")

premium_crops([1500, 2500, 1800, 3200])

```

```python
def system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print(f"Error Count: {errors}")
    print(f"System Status: {status}")

system_health(7)
```

```python
def transaction_check(amount):
    status = "Approved" if amount <= 50000 else "Rejected"
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

transaction_check(60000)
```

```python
def attendance_checker(attendance):
    percentage = (sum(attendance) / len(attendance)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")

attendance_checker([1, 1, 1, 0, 1])

```

```python
def electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print(f"Total Bill: ₹{bill}")
    print(f"Usage Level: {usage}")

electricity_bill(250)
```
