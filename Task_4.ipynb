```python
def mobile_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.")
        elif amount not in valid_plans:
            print("Invalid plan. Available plans:", valid_plans)
        else:
            print("Recharge Successful for ₹", amount)
            break
```

```python
def inventory_alert(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")
```

```python
products = {
    "Rice": 10,
    "Sugar": 20,
    "Oil": 5,
    "Wheat": 30
}

inventory_alert(products)
```

```python
def student_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"
```

```python
marks = [45, 60, 70, 55]
print("Result:", student_result(marks))
```

```python
def cab_fare():
    while True:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        base_fare = 50
        fare = base_fare + (distance * 12)

        if peak == "yes":
            fare += fare * 0.25

        print("Total Fare: ₹", fare)

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break
```

```python
def attendance_eligibility(attendance):
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
```

```python
attendance = ["P", "P", "A", "P", "P", "A", "P"]
print(attendance_eligibility(attendance))
```

```python
def password_strength(password):
    if len(password) < 8:
        return "Weak Password"

    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
```

```python
pwd = input("Enter password: ")
print(password_strength(pwd))
```
