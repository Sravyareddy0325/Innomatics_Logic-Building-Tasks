```python
def parking_lot_manager(capacity, logs):
    parked = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")


# Sample Input
parking_lot_manager(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
def delivery_time_estimator(distance, traffic, weather):
    time = distance * 5  # base time

    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10

    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
delivery_time_estimator(8, "High", "Rainy")
```

```python
def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")


# Sample Input
seat_occupancy(200, [1]*150)
```

```python
def server_load_classifier(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", int(avg_cpu), "%")

    if avg_cpu < 50:
        print("Server Status: Normal")
    elif avg_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
server_load_classifier([45, 60, 70, 85, 90])
```

```python
def resource_usage_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", "Yes" if overused else "No")


# Sample Input
resource_usage_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
def event_registration_controller(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlist = max(0, registrations - capacity)

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status: Closed" if registrations >= capacity else "Open")


# Sample Input
event_registration_controller(100, 105)
```
